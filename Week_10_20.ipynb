```python
from typing import List, Tuple

Point = Tuple[int, int]


# ------------------ GEOMETRY HELPERS ------------------

def orientation(p: Point, q: Point, r: Point) -> int:
    """
    Returns:
        0 -> collinear
        1 -> clockwise
        2 -> counterclockwise
    """
    val = (q[1] - p[1]) * (r[0] - q[0]) - (q[0] - p[0]) * (r[1] - q[1])
    if val == 0:
        return 0
    return 1 if val > 0 else 2


def on_segment(p: Point, q: Point, r: Point) -> bool:
    """Check if point q lies on segment pr"""
    return (min(p[0], r[0]) <= q[0] <= max(p[0], r[0]) and
            min(p[1], r[1]) <= q[1] <= max(p[1], r[1]))


def segments_intersect(p1: Point, q1: Point, p2: Point, q2: Point) -> bool:
    """Check if two line segments intersect"""
    o1 = orientation(p1, q1, p2)
    o2 = orientation(p1, q1, q2)
    o3 = orientation(p2, q2, p1)
    o4 = orientation(p2, q2, q1)

    # General case
    if o1 != o2 and o3 != o4:
        return True

    # Special cases
    if o1 == 0 and on_segment(p1, p2, q1): return True
    if o2 == 0 and on_segment(p1, q2, q1): return True
    if o3 == 0 and on_segment(p2, p1, q2): return True
    if o4 == 0 and on_segment(p2, q1, q2): return True

    return False


# ------------------ POLYGON VALIDITY ------------------

def is_simple_polygon(polygon: List[Point]) -> bool:
    """Check if polygon has no self-intersections"""
    n = len(polygon)

    for i in range(n):
        for j in range(i + 1, n):
            p1, q1 = polygon[i], polygon[(i + 1) % n]
            p2, q2 = polygon[j], polygon[(j + 1) % n]

            # Skip adjacent edges
            if i == j:
                continue
            if (i + 1) % n == j or (j + 1) % n == i:
                continue

            if segments_intersect(p1, q1, p2, q2):
                return False

    return True


# ------------------ POINT IN POLYGON ------------------

def point_in_polygon(point: Point, polygon: List[Point]) -> str:
    """
    Returns: 'inside', 'outside', or 'boundary'
    Uses Ray Casting Algorithm
    """
    x, y = point
    n = len(polygon)
    inside = False

    for i in range(n):
        p1 = polygon[i]
        p2 = polygon[(i + 1) % n]

        # Check boundary
        if orientation(p1, point, p2) == 0 and on_segment(p1, point, p2):
            return "boundary"

        # Ray casting
        if ((p1[1] > y) != (p2[1] > y)):
            xinters = (y - p1[1]) * (p2[0] - p1[0]) / (p2[1] - p1[1]) + p1[0]
            if x < xinters:
                inside = not inside

    return "inside" if inside else "outside"


# ------------------ MAIN ------------------

if __name__ == "__main__":
    V = int(input())

    polygon = []
    for _ in range(V):
        x, y = map(int, input().split())
        polygon.append((x, y))

    if is_simple_polygon(polygon):
        print("VALID")
    else:
        print("INVALID")

    Q = int(input())

    for _ in range(Q):
        x, y = map(int, input().split())
        print(point_in_polygon((x, y), polygon))
```
