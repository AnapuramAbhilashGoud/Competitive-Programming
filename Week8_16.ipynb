```python
from collections import defaultdict, deque


def topological_sort(n, adj):
    """
    Kahn's Algorithm for Topological Sorting
    """
    indegree = [0] * (n + 1)

    for u in adj:
        for v in adj[u]:
            indegree[v] += 1

    queue = deque()
    for i in range(1, n + 1):
        if indegree[i] == 0:
            queue.append(i)

    topo = []

    while queue:
        u = queue.popleft()
        topo.append(u)

        for v in adj[u]:
            indegree[v] -= 1
            if indegree[v] == 0:
                queue.append(v)

    return topo


def longest_path_dag(n, edges, source):
    """
    Find longest path from source in DAG
    """
    adj = defaultdict(list)

    for u, v in edges:
        adj[u].append(v)

    topo = topological_sort(n, adj)

    # Initialize distances
    dist = [-float('inf')] * (n + 1)
    dist[source] = 0

    for u in topo:
        if dist[u] != -float('inf'):
            for v in adj[u]:
                if dist[v] < dist[u] + 1:
                    dist[v] = dist[u] + 1

    return max(dist)


# ------------------ MAIN ------------------
if __name__ == "__main__":
    n, m, s = map(int, input().split())

    edges = []
    for _ in range(m):
        u, v = map(int, input().split())
        edges.append((u, v))

    result = longest_path_dag(n, edges, s)
    print(f"Longest Path Length from {s} = {result}")
```
